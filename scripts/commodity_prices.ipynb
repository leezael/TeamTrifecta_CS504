```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #library used to create visualizations
import seaborn as sns
import datetime
import csv
```

```python
#read file
filename = 'data/commodity_prices.csv'
df = pd.read_csv(filename, encoding = "ISO-8859-1")
df["Date"] = pd.to_datetime(df["Date"], utc=False)
df
```

```python
df['Year'] = pd.DatetimeIndex(df['Date']).year
df
```

```python
df["Aluminium"] = None
df["Aluminium"][df["Commodity"].str.contains("Aluminium")] = "Aluminium"

df["Barley"] = None
df["Barley"][df["Commodity"].str.contains("Barley")] = "Barley"

df["Iron"] = None
df["Iron"][df["Commodity"].str.contains("IronOre")] = "IronOre"

df["CastIron"] = None
df["CastIron"][df["Commodity"].str.contains("CastIron")] = "CastIron"

df["Coal"] = None
df["Coal"][df["Commodity"].str.contains("Coal")] = "Coal"

df["Copper"] = None
df["Copper"][df["Commodity"].str.contains("Copper")] = "Copper"

df["Corn"] = None
df["Corn"][df["Commodity"].str.contains("Corn")] = "Corn"

df["CrudePetroleum"] = None
df["CrudePetroleum"][df["Commodity"].str.contains("CrudePetroleum")] = "CrudePetroleum"

df["Gas"] = None
df["Gas"][df["Commodity"].str.contains("Gas")] = "Gas"

df["Nickel"] = None
df["Nickel"][df["Commodity"].str.contains("Nickel")] = "Nickel"

df["RapeseedOil"] = None
df["RapeseedOil"][df["Commodity"].str.contains("RapeseedOil")] = "RapeseedOil"

df["RefinedPetroleum"] = None
df["RefinedPetroleum"][df["Commodity"].str.contains("RefinedPetroleum")] = "RefinedPetroleum"

df["Slag"] = None
df["Slag"][df["Commodity"].str.contains("Slag")] = "Slag"

df["StainlessSteel"] = None
df["StainlessSteel"][df["Commodity"].str.contains("StainlessSteel")] = "StainlessSteel"

df["Steel"] = None
df["Steel"][df["Commodity"].str.contains("Steel")] = "Steel"

df["SunflowerOil"] = None
df["SunflowerOil"][df["Commodity"].str.contains("SunflowerOil")] = "SunflowerOil"

df["Wheat"] = None
df["Wheat"][df["Commodity"].str.contains("Wheat")] = "Wheat"

print(df)
```

```python
df
df["Date"] = df["Date"].dt.strftime('%m/%Y')
```

```python
nan_value = float("NaN")
df.dropna(how='all',axis=1,inplace=True)
df = df.fillna('')
df.replace('', np.nan, inplace=True)
df.dropna()
display(df)
```
